```python
import re

def get_furnishing_count(details, furnishing):
    if isinstance(details, str):
        if f"No {furnishing}" in details:
            return 0
        pattern = re.compile(rf"(\d+)\s+{re.escape(furnishing)}", re.IGNORECASE)
        match = pattern.search(details)
        if match:
            return int(match.group(1))
        elif furnishing.lower() in details.lower():
            return 1
    return 0
```

```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
import ast
import re

# Step 1: Load data


# Step 2: Clean and parse furnishDetails
def parse_furnish_details(val):
    if pd.notnull(val) and val.startswith('['):
        try:
            items = ast.literal_eval(val)
            return [
                re.sub(r'^\d+\s+', '', item).strip()
                for item in items
                if isinstance(item, str) and not item.startswith('No')
            ]
        except:
            return []
    return []

df['furnish_list'] = df['furnishDetails'].apply(parse_furnish_details)

# Step 3: Binarize
mlb = MultiLabelBinarizer()
furnish_matrix = mlb.fit_transform(df['furnish_list'])
furnish_df = pd.DataFrame(furnish_matrix, columns=mlb.classes_)

# Step 4: Define weights
weights = {
    'AC': 3,
    'Bed': 2,
    'Chimney': 1,
    'Curtains': 1,
    'Dining Table': 2,
    'Exhaust Fan': 1,
    'Fan': 1,
    'Fridge': 2,
    'Geyser': 2,
    'Light': 1,
    'Microwave': 2,
    'Modular Kitchen': 3,
    'Sofa': 2,
    'Stove': 1,
    'TV': 2,
    'Wardrobe': 2,
    'Washing Machine': 2,
    'Water Purifier': 2
}

# Step 5: Compute luxury score
available_items = [item for item in weights.keys() if item in furnish_df.columns]
df['luxury_score'] = furnish_df[available_items].multiply(
    [weights[i] for i in available_items], axis=1
).sum(axis=1)

# Optional: check
print(df[['furnishDetails', 'luxury_score']].head())
```

```python
def scorer(model_name,model):
    output = []
    output.append(model_name)
    pipeline=Pipeline([
    ('preprocessor',preprocessor),
    ('regressor',model)
    ])
    kfold=KFold(n_splits=10,shuffle=True,random_state=42)
    scores=cross_val_score(pipeline,X,y_transformed,cv=kfold,scoring='r2')
    output.append(scores.mean())
    x_train,x_test,y_train,y_test = train_test_split(X,y_transformed,test_size=0.2,random_state=42)
    pipeline.fit(x_train,y_train)
    y_pred=pipeline.predict(x_test)
    y_pred = np.expm1(y_pred)
    output.append(mean_absolute_error(np.expm1(y_test),y_pred))
    return output


```
